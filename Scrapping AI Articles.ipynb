```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
web=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
```

```python
web
```

```python
soup=BeautifulSoup(web.content,'html.parser')
```

```python
articles_data=soup.find("div", class_="sc-orwwe2-3 jOMrrY")
```

```python
articles_data.prettify()
```

```python
titles = []
authors = []
dates = []
urls = []
```

```python
for article in articles_data.find_all("li"):
    title = article.find("h2").text.strip()
    titles.append(title)
  
```

```python
for article in articles_data.find_all("li"):
    
    author = article.find("span", class_="sc-1w3fpd7-0 dnCnAO").text.strip()
    authors.append(author)
  
```

```python
for article in articles_data.find_all("li"):
    date = article.find("span", class_="sc-1w3fpd7-0 dnCnAO").text.strip()
    dates.append(date)
```

```python
for article in articles_data.find_all("li"):
    url = article.find("a")["href"]
    urls.append(url)
```

```python
data = {
  "Paper Title": titles,
  "Authors": authors,
  "Published Date": dates,
  "Paper URL": urls
}
```

```python
df=pd.DataFrame(data)
```
